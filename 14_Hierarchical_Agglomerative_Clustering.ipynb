```python
df=pd.read_csv("CC GENERAL.csv")
df.head(3)
```

```python
df.drop('CUST_ID',axis=1,inplace=True)
df.head(3)
```

```python
# print(df.isnull().sum())
df.fillna(df.mean(),inplace=True)
df.isnull().sum()
```

```python
# Manual scaling

for column in df.columns:
    df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())
df.head(3)
```

*QUESTION : Why is scaling required for distance-based algorithms?*

*ANSWER : large-valued features dominate the distance and distort clustering → scaling fixes this.*

## PART 2

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
wcss = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(df)
    wcss.append(km.inertia_)

plt.plot(range(2, 11), wcss, marker='o')
plt.xlabel("K")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

*Decrease slows down at 2 or 7*

```python
from sklearn.metrics import silhouette_score

scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(df)
    sil = silhouette_score(df, labels)
    scores.append(sil)

plt.plot(range(2, 11), scores, marker='o')
plt.title("Silhouette Scores")
plt.xlabel("K")
plt.ylabel("Score")
plt.show()
```

*Best Score = 2-> 0.38*

## PART 3


Manual Implementation

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

def maxDistanceClusters(a,b,df):
    dist=0
    for i in a:
        for j in b:
            d=euclidean_distance(df.iloc[i].values,df.iloc[j].values)
            # d=euclidean_distance(df[i],df[j])
            if d>dist:
                dist=d
    return dist
```

```python
def agglomerativeClustering(df, num_clusters):
    clusters = [[i] for i in range(len(df))]
    
    while len(clusters) > num_clusters:
        max_dist = -1
        to_merge = (0, 0)
        
        for i in range(len(clusters)):
            for j in range(i + 1, len(clusters)):
                dist = maxDistanceClusters(clusters[i], clusters[j], df)
                if dist > max_dist:
                    max_dist = dist
                    to_merge = (i, j)
        
        i, j = to_merge
        clusters[i].extend(clusters[j])
        del clusters[j]
    
    return clusters

df_sample = df.sample(200, random_state=42).reset_index(drop=True)
clusters = agglomerativeClustering(df_sample, 2)
print("Clusters formed:", clusters)
for idx, cluster in enumerate(clusters):
    print(f"Cluster {idx + 1}: {cluster}")
```

Sklearn Implementation

```python
from sklearn.cluster import AgglomerativeClustering


agg_sklearn = AgglomerativeClustering(n_clusters=2, linkage='complete')
labels_sklearn = agg_sklearn.fit_predict(df_sample)

# Evaluate
sil_score = silhouette_score(df_sample, labels_sklearn)
print(f"Sklearn Agglomerative Clustering (Complete Linkage)")
print(f"Silhouette Score: {sil_score:.3f}")

# Show cluster sizes
unique, counts = np.unique(labels_sklearn, return_counts=True)
for cluster_id, count in zip(unique, counts):
    print(f"Cluster {cluster_id}: {count} samples")
```

```python
# K-Means on same sample
kmeans = KMeans(n_clusters=2, random_state=42)
labels_kmeans = kmeans.fit_predict(df_sample)

# Agglomerative (already computed above)
labels_agg = labels_sklearn

# Compare Silhouette Scores
sil_kmeans = silhouette_score(df_sample, labels_kmeans)
sil_agg = silhouette_score(df_sample, labels_agg)

print("COMPARISON: K-Means vs Agglomerative Clustering")
print()
print(f"K-Means Silhouette Score: {sil_kmeans:.3f}")
print(f"Agglomerative Silhouette Score: {sil_agg:.3f}")
print()

# Cluster sizes comparison
print("K-Means Cluster Sizes:")
unique_km, counts_km = np.unique(labels_kmeans, return_counts=True)
for cluster_id, count in zip(unique_km, counts_km):
    print(f"  Cluster {cluster_id}: {count} samples")

print("\nAgglomerative Cluster Sizes:")
unique_agg, counts_agg = np.unique(labels_agg, return_counts=True)
for cluster_id, count in zip(unique_agg, counts_agg):
    print(f"  Cluster {cluster_id}: {count} samples")
```
