```python
playerInjury=[
    [1,"Goalkeeper","Low","Head"],
    [2,"Defender","High","Leg"],
    [3,"Forward","Medium","Leg"],
    [4,"Defender","Medium","Shoulder"],
    [5,"Forward","High","Head"]
]
def chi_square_test(table):
    rows = len(table)
    cols =len(table[0])

    total =sum(sum(row) for row in table)
    row_totals =[sum(row) for row in table]

    col_totals =[sum(table[r][c] for r in range(rows)) for c in range(cols)]
    chi2 = 0


    for r in range(rows):
        for c in range(cols):
            expected =(row_totals[r] * col_totals[c]) / total
            observed =table[r][c]
            chi2 +=(observed - expected) ** 2 / expected if expected != 0 else 0


    return chi2
```

```python
# TASK A: Position effects Injury?
GH=0
GL=0
GS=0
DH=0
DL=0
DS=0
FH=0
FL=0
FS=0
n=len(playerInjury)
for i in range (n):
    if playerInjury[i][1]=="Goalkeeper" and playerInjury[i][3]=="Head":
        GH+=1
    elif playerInjury[i][1]=="Goalkeeper" and playerInjury[i][3]=="Leg":
        GL+=1
    elif playerInjury[i][1]=="Goalkeeper" and playerInjury[i][3]=="Shoulder":
        GS+=1
    elif playerInjury[i][1]=="Defender" and playerInjury[i][3]=="Head":
        DH+=1
    elif playerInjury[i][1]=="Defender" and playerInjury[i][3]=="Leg":
        DL+=1
    elif playerInjury[i][1]=="Defender" and playerInjury[i][3]=="Shoulder":
        DS+=1
    elif playerInjury[i][1]=="Forward" and playerInjury[i][3]=="Head":
        FH+=1
    elif playerInjury[i][1]=="Forward" and playerInjury[i][3]=="Leg":
        FL+=1
    elif playerInjury[i][1]=="Forward" and playerInjury[i][3]=="Shoulder":
        FS+=1

# print(GH,GL,GS,DH,DL,DS,FH,FL,FS)
injuryTable=[
    [GH,GL,GS],
    [DH,DL,DS],
    [FH,FL,FS]
]
print(injuryTable)
chi2Val=chi_square_test(injuryTable)
print(chi2Val)
if chi2Val < 9.488:
    print("Fail to reject the null hypothesis (Ho): \n   There is no significant association between player position and injury type.")
else:
    print("Reject the null hypothesis (Ho): There is a significant association between player position and injury type.")
```

```python
# TASK B: Training Intensity effects Injury Type?


LH=0
LL=0
LS=0
MH=0
ML=0
MS=0
HH=0
HL=0
HS=0
n=len(playerInjury)
for i in range (n):
    if playerInjury[i][2]=="Low" and playerInjury[i][3]=="Head":
        LH+=1
    elif playerInjury[i][2]=="Low" and playerInjury[i][3]=="Leg":
        LL+=1
    elif playerInjury[i][2]=="Low" and playerInjury[i][3]=="Shoulder":
        LS+=1
    elif playerInjury[i][2]=="Medium" and playerInjury[i][3]=="Head":
        MH+=1
    elif playerInjury[i][2]=="Medium" and playerInjury[i][3]=="Leg":
        ML+=1
    elif playerInjury[i][2]=="Medium" and playerInjury[i][3]=="Shoulder":
        MS+=1
    elif playerInjury[i][2]=="High" and playerInjury[i][3]=="Head":
        HH+=1
    elif playerInjury[i][2]=="High" and playerInjury[i][3]=="Leg":
        HL+=1
    elif playerInjury[i][2]=="High" and playerInjury[i][3]=="Shoulder":
        HS+=1

intensityTable=[
    [LH,LL,LS],
    [MH,ML,MS],
    [HH,HL,HS]
]

print(intensityTable)

chi2Val_intensity=chi_square_test(intensityTable)
print(chi2Val_intensity)

# For 3x3 table, df = (3-1)*(3-1) = 4, critical value at α=0.05 is 9.488
if chi2Val_intensity < 9.488:
    print("Fail to reject the null hypothesis (Ho): \n   There is no significant association between training intensity and injury type.")
else:
    print("Reject the null hypothesis (Ho): There is a significant association between training intensity and injury type.")
```

```python
# PART C: Determine strongest association and provide injury prevention recommendation
total_players = len(playerInjury)

position_strength = (chi2Val / (total_players * min(len(injuryTable)-1, len(injuryTable[0])-1)))**(0.5)
print("Player Position and Injury Type Strength : ", position_strength)

intensity_strength = (chi2Val_intensity / (total_players * min(len(intensityTable)-1, len(intensityTable[0])-1)))**(0.5)
print("Training Intensity and Injury Type Strength : ", intensity_strength)

if position_strength > intensity_strength:
    print("Association of Player Position to Injury Type is stronger than Training Intensity to Injury Type.")
elif position_strength == intensity_strength:
    print("Both associations are equally strong. Both have strength = ", position_strength)
else:
    print("Association of Training Intensity to Injury Type is stronger than Player Position to Injury Type.")
```
