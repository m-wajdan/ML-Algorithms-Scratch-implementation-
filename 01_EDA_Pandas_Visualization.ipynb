```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Pokemon.csv")
```

```python
# 1. Load the Pokémon dataset and display the first 10 rows.
df.head(10)
```

```python
# 2. Find the number of Pokémon and total features (columns) in the dataset.
num_pokemon = df.shape[0]
print(num_pokemon)
num_features = df.shape[1]
print(num_features)
```

```python
num_pokemon = df['Name'].nunique()
print("Number of unique Pokémon:", num_pokemon)
for i in df['Name'].unique():
    print(i)

num_features = df.count(axis=1).shape[0]
print("Total features (columns):", df.shape[1])
```

```python
# Show the data types of each column and check for missing values.
print(df.dtypes)
print("==================")
print(df.isnull().sum())
```

```python
# 4. Print the unique values of Type 1 and count how many Pokémon belong to each type.
print(df["Type 1"].unique())
print("=================")
print(df["Type 1"].value_counts())
```

```python
# 5. Calculate the average HP of all Pokémon.
print(df["HP"].mean())
```

```python
# 6. Find the Pokémon with the highest Attack and the one with the lowest Defense.
df.sort_values(by="Attack", ascending=False, inplace=True)
print("Pokémon with the highest Attack:"+df.iloc[0]["Name"])
print("Pokémon with the lowest Attack:"+df.iloc[-1]["Name"])
```

```python
df.sort_values(by="Defense", ascending=True, inplace=True)
print("Pokémon with the highest Defense:"+df.iloc[0]["Name"])
print("Pokémon with the lowest Defense:"+df.iloc[-1]["Name"])
```

```python
# 7. Compare the average Speed of Legendary vs Non-Legendary Pokémon.
legendaryAvg=df[df["Legendary"]==True]["Speed"].mean()
print(legendaryAvg)
nonLegendaryAvg=df[df["Legendary"]==False]["Speed"].mean()
print(nonLegendaryAvg)
```

```python
print(df.groupby("Legendary")["Speed"].mean())
```

```python
# Create a new column Total Stats = HP + Attack + Defense + Sp. Atk + Sp. Def + Speed.
df["Total Stats"]=df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"]
df.head(3)
```

```python
# 9. Find the top 5 Pokémon with the highest Total Stats.
df.sort_values(by="Total Stats",ascending=False)
df.head(5)
```

```python
# 10. Group Pokémon by Generation and calculate the average Attack in each generation.
df.groupby("Generation")["Attack"].mean()
```

```python
# 11. Plot a histogram of HP values.
sns.histplot(df["HP"], bins=20)
```

```python
# 12. Create a bar chart showing the count of Pokémon for each Type 1.
plt.bar(df["Type 1"].value_counts().index, df["Type 1"].value_counts())
plt.xticks(rotation=90)
plt.xlabel("Type 1")
plt.ylabel("Count")
plt.show()
```

```python
# 13. Make a scatter plot of Attack vs Defense, coloring points by Legendary status.
plt.figure(figsize=(8,6))
colors = {True: 'red', False: 'blue'}
plt.scatter(df['Attack'], df['Defense'], c=df['Legendary'].map(colors), label=None)
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack vs Defense (Legendary in Red)')
plt.show()
```

```python
# 14. Draw a boxplot comparing Speed of Legendary vs Non-Legendary Pokémon.
sns.boxplot(x='Legendary', y='Speed', data=df)
```

```python
# 15. Create a heatmap of correlations between numerical features (HP, Attack, etc.).
corr = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Total Stats']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Pokémon Stats')
```
