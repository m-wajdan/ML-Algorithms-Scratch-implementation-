```python
import numpy as np
import pandas as pd

data=[
    [1,22],
    [2,27],
    [3,34],
    [4,41],
    [5,43]
]

X=[row[0] for row in data]
y=[row[1] for row in data]
print("X:", X)
print("y:", y)
```

```python
# Manually calculate the following using basic Python operations:
# ● Mean of X and Y
# ● Slope (m)
# ● Intercept (c)

meanX=sum(X)/len(X)
meanY=sum(y)/len(y)
print("Mean of X:", meanX)
print("Mean of y:", meanY)
slopeNum=sum((X[i]-meanX)*(y[i]-meanY) for i in range(len(X)))
slopeDen=sum((X[i]-meanX)**2 for i in range(len(X)))
slope=slopeNum/slopeDen
print("Slope (m):", slope)
intercept=meanY - slope*meanX
print("Intercept (c):", intercept)
```

```python
# Construct the regression line using the equation y=mx+c.
print(f"Regression Equation : y = {slope} x + {intercept}")

# Predict values of Y using your calculated slope and intercept.
X=np.array(X)
yPred=slope*X+intercept
print(yPred)

#Mean Squared Error:
y=np.array(y)
mse=np.mean((y-yPred)**2)
print("MES = ", mse)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Actual values')
plt.plot(X, yPred, color='red', linewidth=2, label='Predicted (Regression Line)')
plt.title('Actual vs Predicted Values')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Task 2: Linear Regression using Built-in Python Library
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = np.array([1,2,3,4,5])
Y = np.array([22,27,34,41,43])

X = np.array(X).reshape(-1, 1)

model = LinearRegression()
model.fit(X,Y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print(f"Slope (coefficient): {slope}")
print(f"Intercept: {intercept}")
```

```python
TestSample=np.array([10, 15]).reshape(-1, 1)
yPred = model.predict(TestSample)
print("Predictions : ", yPred)
```

```python
Y_pred_train = model.predict(X)
mse = mean_squared_error(Y, Y_pred_train)
r2 = r2_score(Y, Y_pred_train)

print("MSE : ", mse)
print("R squared : ", r2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Experience': [1, 2, 3, 4, 5],
    'Age': [22, 25, 28, 35, 40],
    'TestScore': [50, 60, 70, 80, 85],
    'HighSalary': [0, 0, 1, 1, 1]
}
```

```python
df = pd.DataFrame(data)
print("Dataset:")
df
```

```python
X = df[['Experience', 'Age', 'TestScore']].values
y = df['HighSalary'].values
print("Features (X):")
print(X)
print("\nLabels (y):")
print(y)
```

```python
weights = np.array([0.0, 0.0, 0.0])
bias = 0.0
learning_rate = 0.01
```

```python
import math

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# Binary Cross-Entropy Loss tells how wrong the predictions are
def binary_cross_entropy_loss(y_true, y_pred):
    loss = 0
    for yt, yp in zip(y_true, y_pred):
        loss += - (yt * math.log(yp + 1e-9) + (1 - yt) * math.log(1 - yp + 1e-9))
    return loss / len(y_true)

# Dataset
X = [
    [1, 22, 51],
    [2, 25, 60],
    [3, 28, 70],
    [4, 35, 80],
    [5, 40, 85]
]
y = [0, 0, 1, 1, 1]

# Initial parameters 
weights = [0.0, 0.0, 0.0]
bias = 0.0
learning_rate = 0.001   
iterations = 50

# Training Loop
# learns the weights 
#
for it in range(1, iterations + 1):
    y_pred = []
    
    # Forward pass
    for features in X:
        z = sum(w * x for w, x in zip(weights, features)) + bias  
        y_pred.append(sigmoid(z))

    # Compute loss
    loss = binary_cross_entropy_loss(y, y_pred)

    # Gradients
    dw = [0.0, 0.0, 0.0]
    db = 0.0
    for i in range(len(X)):
        error = y_pred[i] - y[i]
        for j in range(len(weights)):
            dw[j] += error * X[i][j]
        db += error

    # Gradient descent update
    for j in range(len(weights)):
        weights[j] -= learning_rate * (dw[j] / len(X))
    bias -= learning_rate * (db / len(X))

    # Print every 10 iterations
    if it % 10 == 0:
        print(f"Iteration {it}")
        print(f"Loss: {loss:.4f}")
        print(f"Weights: {weights}")
        print(f"Bias: {bias:.4f}")
        print("-" * 40)

print("\nFinal Predictions (after training):")
for features in X:
    z = sum(w * x for w, x in zip(weights, features)) + bias
    prob = sigmoid(z)
    pred = 1 if prob >= 0.5 else 0
    print(f"Input: {features} → Predicted Probability: {prob:.4f} → Class: {pred}")
```
