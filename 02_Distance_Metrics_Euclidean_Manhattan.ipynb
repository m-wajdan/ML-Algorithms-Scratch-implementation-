```python
# Customer = [Groceries, Electronics, Clothing]
Data = [
    [200, 100, 150],
    [220, 90, 160],
    [180, 400, 170],
    [210, 95, 155],
    [190, 110, 145]
]
list_of_avgs = []

for i in range(3):  
    avgs = 0
    for j in range(5):  
        avgs += Data[j][i]
    avg = avgs / 5  
    list_of_avgs.append(avg)
print(list_of_avgs)


def l1_distance(customer_data, averages):
    distance = 0
    for i in range(len(customer_data)):
        distance += abs(customer_data[i] - averages[i])
    return distance
distances=[]
for index, customer in enumerate(Data):
    distance = l1_distance(customer, list_of_avgs)
    distances.append(distance)
    print(f"L1 distance for Customer {index + 1}: {distance}")
```

```python
max_dist = -1
max_index = -1
for i in range(len(distances)):
    if distances[i] > max_dist:
        max_dist = distances[i]
        max_index = i

customers = ["C1", "C2", "C3", "C4", "C5"]
outlier_customer = customers[max_index]

print("Outlier Customer:", outlier_customer)
```

```python
sorted_dist = sorted(distances)
n = len(sorted_dist)



q1_index = n // 4
Q1 = sorted_dist[q1_index]

q3_index = (3 * n) // 4
Q3 = sorted_dist[q3_index]

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = []
for i in range(n):
    if distances[i] < lower_bound or distances[i] > upper_bound:
        outliers.append(customers[i])

print("Q1:", Q1, "Q3:", Q3, "IQR:", IQR)
print("Lower Bound:", lower_bound, "Upper Bound:", upper_bound)
print("Outlier Customers:", outliers)
```
