```python
x=[3,4]
u=[1,2]
# proj of x on u
```

```python
XdotU=0
UdotU=0
for i in range (len(x)):
    XdotU+=(x[i]*u[i])
    UdotU+=(u[i]*u[i])
print(XdotU)
print(UdotU)
```

```python
projection=[(XdotU/UdotU)*u_i for u_i in u]
print(projection)
```

```python
import numpy as np

x = np.array([3, 4])
u = np.array([1, 2])

# Projection of x on u
x_dot_u = x.dot(u)
u_dot_u = u.dot(u)

projection = (x_dot_u / u_dot_u) * u
print("Projection:", projection)
```

```python
for i in dict(counts):
    print("prob of grade ",i," is ",counts[i]/len(grades))
```

```python
sumOfProb=0
for i in dict(counts):
    sumOfProb+=counts[i]/len(grades)
print("Sum of all probabilities is ",sumOfProb)
```

# Question # 3

```python
Data = [("Male", "Pass"), ("Female", "Pass"), ("Male", "Fail"),("Female", "Fail"), ("Male", "Pass"), ("Female", "Pass")]
```

```python
MP=0
MF=0
FP=0
FF=0
n=len(Data)
for i in range (len(Data)):
    if Data[i][0]=="Male" and Data[i][1]=="Pass":
        MP+=1
    elif Data[i][0]=="Female" and Data[i][1]=="Pass":
        FP+=1
    elif Data[i][0]=="Male" and Data[i][1]=="Fail":
        MF+=1
    else:
        FF+=1
# print(MP,MF,FP,FF)
jointPMF=[
    [MP/n,MF/n],
    [FP/n,FF/n],
    [(MP/n)+(FP/n),(MF/n)+(FF/n)]
]
print("Joint PMF is ")
jointPMF
```

```python
sumOfJointPMF=jointPMF[2][0]+jointPMF[2][1]
print(sumOfJointPMF)
```

# QUESTION # 4

```python
import numpy as np
values = np.random.normal(loc=0, scale=1, size=1000)
print(values[:10])
```

```python
# f(x)=1/underROot(2pi) * e^(-x^2/2)
# pdf_scipy = norm.pdf(x, loc=0, scale=1)
```

```python
pdf_formula = (1/np.sqrt(2*np.pi)) * np.exp(-np.sort(values)**2 / 2)
# print(pdf_formula[:10])
```

```python
plt.plot(np.sort(values), pdf_formula, label="PDF (Formula)", color="blue")
plt.title("Standard Normal Distribution PDF")
plt.xlabel("x")
plt.ylabel("f(x)")  
plt.legend()
plt.grid(True)
plt.show()
```

```python
area_formula = np.trapezoid(pdf_formula, np.sort(values))
print("Area under curve (Formula):", area_formula)
```

# QUESTION # 5

```python
Marks = [35, 40, 55, 60, 72, 80, 85, 90, 95]
```

```python
#Equal WIdth
min_mark = min(Marks)
max_mark = max(Marks)
width = (max_mark - min_mark) / 3

def equal_width_category(mark):
    if mark < min_mark + width:
        return "Low"
    elif mark < min_mark + 2*width:
        return "Medium"
    else:
        return "High"

equalWidth = [equal_width_category(m) for m in Marks]
print(Marks)
print("Equal-width categories:", equalWidth) 
```

```python
sorted_marks = sorted(Marks)
n = len(Marks)
freq = n // 3

ef_categories = [""] * n
for i, mark in enumerate(sorted_marks):
    if i < freq:
        category = "Low"
    elif i < 2*freq:
        category = "Medium"
    else:
        category = "High"
    ef_categories[Marks.index(mark)] = category

print("Equal-frequency categories:", ef_categories)
```

```python

print("\nMark    Equal-width      Equal-frequency")
for i, mark in enumerate(Marks):
    print(f"{mark}\t{equalWidth[i]}\t\t{ef_categories[i]}")
```
